```python
import os
from pathlib import Path
import random
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

import torch 
import torch.nn as nn 
from torch.utils.data import Dataset,DataLoader, random_split
from torchvision import transforms,models
```

```python
data_path = Path("brain_tumor_dataset")
train_path = data_path/"Training"
test_path = data_path/"Testing"
```

```python
def walk_through_directory(data_path):

    for dirpaths,dirnames,filenames in os.walk(data_path):
        print(f"There are {len(dirnames)} directories, {len(filenames)} files in {dirpaths}")
```

```python
def get_class_names(path:str) -> list: 

    class_names = [classes.name for classes in os.scandir(path)]
    class_to_idx = {classes:index for index,classes in enumerate(class_names)}
    idx_to_class = {index:classes for index,classes in enumerate(class_names)}
    return class_names,class_to_idx,idx_to_class
```

```python
class_names,class_to_idx,idx_to_class = get_class_names(train_path)
```

```python
class BrainTumorDataset(Dataset):

    def __init__(self,path:str,transform:None):

        self.paths = list(path.glob("*/*.jpg"))
        self.classes,self.class_to_idx,self.idx_to_class = get_class_names(path)
        self.transform = transform 

    def load_image(self,index:int):
        image_path = self.paths[index]
        img = Image.open(image_path)
        return img.convert("RGB")

    def __len__(self):
        return len(self.paths)

    def __getitem__(self,index):
        img = self.load_image(index)
        class_name = self.paths[index].parent.name
        class_to_idx = self.class_to_idx[class_name]

        if self.transform is None:
            return img,class_to_idx

        transformed_img = self.transform(img)
        return transformed_img,class_to_idx
        
```

```python
train_transforms = transforms.Compose([
    transforms.Resize((224, 224)), 
    transforms.RandomCrop((200, 200)), 
    transforms.Resize((224, 224)),  
    transforms.ToTensor(), 
])

test_transforms = transforms.Compose([
    transforms.Resize((224, 224)),  
    transforms.ToTensor(),
])
```

```python
train_dataset = BrainTumorDataset(train_path,train_transforms)
test_dataset = BrainTumorDataset(test_path,test_transforms)
```

```python
len(train_dataset),len(test_dataset)
```

```python
val_size = int(0.15*len(train_dataset))
train_size = len(train_dataset) - val_size

train_dataset,val_dataset = random_split(train_dataset,[train_size,val_size])
```

```python
len(train_dataset),len(val_dataset),len(test_dataset)
```

```python
train_dataloader = DataLoader(dataset=train_dataset,batch_size=32,num_workers=1,shuffle=True)
val_dataloader = DataLoader(dataset=val_dataset,batch_size=32,num_workers=1,shuffle=False)
test_dataloader = DataLoader(dataset=test_dataset,batch_size=32,num_workers=1,shuffle=False)
```

```python
img,label = next(iter(train_dataloader))
```

```python
plt.imshow(img[0].permute(1,2,0))
plt.title(idx_to_class[label[0].item()])
plt.show()
```

```python
model = models.resnet34(weights=models.ResNet34_Weights.IMAGENET1K_V1)

for param in model.parameters():
    param.requires_grad=False

for param in model.layer4.parameters():
    param.requires_grad=True

model.fc = nn.Linear(model.fc.in_features, len(class_names))
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
x = img[0].unsqueeze(dim=0)
model(x)
```

```python
def accuracy_fn(y_true,y_pred):
    
    correct = torch.eq(y_true,y_pred).sum().item()
    acc = 100*(correct/len(y_pred))
    return acc
```

```python
def train_step(dataloader:DataLoader,train_loss:list,train_acc:list):

    
    epoch_train_loss = 0
    epoch_train_acc = 0

    for X,y in train_dataloader:

        model.train()
        y_pred = model(X)
        tr_loss = loss_fn(y_pred.squeeze(),y)
        tr_acc = accuracy_fn(y,y_pred.argmax(dim=1))

        epoch_train_loss += tr_loss
        epoch_train_acc += tr_acc

        optimizer.zero_grad()
        tr_loss.backward()
        optimizer.step()

    epoch_train_loss/=len(train_dataloader)
    epoch_train_acc/=len(train_dataloader)

    train_loss.append(epoch_train_loss)
    train_acc.append(epoch_train_acc)

    return epoch_train_loss,epoch_train_acc
```

```python
def test_step(dataloader:DataLoader,test_loss:list,test_acc:list):

   

    epoch_test_loss = 0
    epoch_test_acc = 0

    with torch.inference_mode():

        model.eval()
        for X,y in val_dataloader:
             
            y_pred = model(X)
            te_loss = loss_fn(y_pred,y)
            te_acc = accuracy_fn(y,y_pred.argmax(dim=1))
    
            epoch_test_loss += te_loss
            epoch_test_acc += te_acc
    
            
    epoch_test_loss = epoch_test_loss/len(val_dataloader)
    epoch_test_acc = epoch_test_acc/len(val_dataloader)

    test_loss.append(epoch_test_loss)
    test_acc.append(epoch_test_acc)

    return epoch_test_loss,epoch_test_acc
```

```python
def train(epochs,train_dataloader,test_dataloader):

    train_loss = []
    train_acc = []
    test_loss = []
    test_acc = []
    
    for epoch in range(epochs):

        epoch_train_loss,epoch_train_acc = train_step(train_dataloader,train_loss,train_acc)
        epoch_test_loss,epoch_test_acc = test_step(test_dataloader, test_loss, test_acc)

        print(f"\n{epoch+1} Train loss: {epoch_train_loss} | Train Accuracy: {epoch_train_acc} | Validation Loss: {epoch_test_loss} | Validation Acc: {epoch_test_acc}")


    return train_loss,train_acc,test_loss,test_acc
```

```python
train_loss,train_acc,test_loss,test_acc = train(10,train_dataloader,val_dataloader)
```

```python
def evaluate(test_dataloader):

    test_acc = 0 

    for X,y in test_dataloader:
            y_pred = model(X)
            test_acc+= accuracy_fn(y,y_pred.argmax(dim=1))

    return test_acc/len(test_dataloader)
```

```python
test_acc = evaluate(test_dataloader)
```
